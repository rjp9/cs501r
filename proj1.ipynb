```python
from PIL import Image, ImageDraw, ImageFont
import IPython.display
from random import randint as rnd

X_DIM = 512
Y_DIM = 288

greetings = ["hey", "yo", "what's up?", "hi", "hello", 
            "henlo", "greetings", "sup", "Ayy!", "uhh"]

def show(img):
  IPython.display.display(img)
  
img = Image.new("RGBA", (512, 288), "black")
txt = ImageDraw.Draw(img)

for i in range(50):
  coords = (rnd(0, X_DIM - 100), rnd(0, Y_DIM - 20))  
  color = (rnd(0, 255), rnd(0, 255), rnd(0, 255))
  text = greetings[rnd(0, len(greetings)-1)]
  size = rnd(10,40)
  fnt = ImageFont.truetype('LDFComicSansBold.ttf', size)
  #fnt= None
  # print(coords, color)
  txt.text(coords, text, font=fnt, fill=color)

show(img)
```
